#  Plot the frequence of the different incidents whith impact on punctuality

```python
# Calculate the frequency of each modality
incident_counts = df_incident_impact_punc["Description de l'incident.2"].value_counts()
sns.barplot(x=incident_counts.values, y=incident_counts.index, hue=incident_counts.index, palette='viridis')
plt.title('Incidents causing punctiality problem')
plt.xlabel('Frequency')
plt.ylabel('Incident')
plt.show()
```

```python
# =============================================================================
#  Plot the sum of minutes delay for the different incident category
# =============================================================================
delay_sums_by_incident = df_incident_impact_punc.groupby("Description de l'incident.2")["Nombre de minutes de retard"].sum().reset_index()
sns.barplot(x="Nombre de minutes de retard", y="Description de l'incident.2", hue = "Description de l'incident.2", data=delay_sums_by_incident, palette='viridis')
plt.title('Sum of minutes delay for Different incidents')
plt.xlabel('Sum of minutes delay')
plt.ylabel('Incident')
plt.show()
```

```python
# Box Plot
sns.boxplot(x="Nombre de minutes de retard", y="Description de l'incident.2", data= df_incident_impact_punc)
plt.title('Box Plot delay per incident')
```

### Invistigate the departure delay of train in term of Relation 

```python
np.unique(df_punc_d_1['Relation'])
```

```python
df_punc_d_1.columns
```

```python
# Convert the departure and arrival delay to minute
df_punc_d_1["Retard à l'arrivée"] = df_punc_d_1["Retard à l'arrivée"]/60
df_punc_d_1["Retard au départ"] = df_punc_d_1["Retard au départ"]/60
```

```python
# Departure delay in term of relation 
delay_sums_dep_by_relation = df_punc_d_1.groupby('Relation')['Retard au départ'].sum().reset_index()
delay_sums_dep_by_relation_IC = delay_sums_dep_by_relation[delay_sums_dep_by_relation['Relation'].str.startswith('IC')]
delay_sums_dep_by_relation_L = delay_sums_dep_by_relation[delay_sums_dep_by_relation['Relation'].str.startswith('L ')]
delay_sums_dep_by_relation_Other = delay_sums_dep_by_relation[~delay_sums_dep_by_relation['Relation'].str.startswith('IC') & ~delay_sums_dep_by_relation['Relation'].str.startswith('L ')]
```

```python
# Plot for the departure delay in term of relation IC
delay_sums_dep_by_relation_IC = delay_sums_dep_by_relation_IC.sort_values('Retard au départ', ascending=False)
sns.barplot(x='Retard au départ', y='Relation', hue = 'Relation', data=delay_sums_dep_by_relation_IC, palette='viridis')
plt.title('Sum of seconds delay departure for Different Relation')
plt.xlabel('Sum of seconds delay')
plt.ylabel('Relation')
plt.show()
```

```python
# Plot for the departure delay in term of relation L 
delay_sums_dep_by_relation_L = delay_sums_dep_by_relation_L.sort_values('Retard au départ', ascending=False)
# Select only the relation above 500 mins delay to have a better visualisation
delay_sums_dep_by_relation_L_higher_500 = delay_sums_dep_by_relation_L[delay_sums_dep_by_relation_L['Retard au départ'] > 500]
sns.barplot(x='Retard au départ', y='Relation', hue = 'Relation', data=delay_sums_dep_by_relation_L_higher_500, palette='viridis')
plt.title('Sum of seconds delay departure for Different Relation')
plt.xlabel('Sum of seconds delay')
plt.ylabel('Relation')
plt.show()
```

```python
# Plot for the departure delay in term of relation Europe and Other
delay_sums_dep_by_relation_Other = delay_sums_dep_by_relation_Other.sort_values('Retard au départ', ascending=False)
sns.barplot(x='Retard au départ', y='Relation', hue = 'Relation', data=delay_sums_dep_by_relation_Other, palette='viridis')
plt.title('Sum of seconds delay departure for Different Relation')
plt.xlabel('Sum of seconds delay')
plt.ylabel('Relation')
plt.show()
```

```python
# How much total delay cause each kind of ETCS 
df_merged_delay_per_etcs = df_merged_delay_etcs.groupby("Niveau d'ETCS")["Retard à l'arrivée"].sum().sort_values(ascending = False)
df_merged_delay_per_etcs = df_merged_delay_per_etcs.reset_index()
df_merged_delay_per_etcs
```

```python
# How much Percentage delay cause each kind of ETCS over the total delay
rate_delay_per_ETCS_Cat = df_merged_delay_per_etcs["Retard à l'arrivée"]/sum(df_merged_delay_per_etcs["Retard à l'arrivée"])
rate_delay_per_ETCS_Cat
```

```python
# Ratio of delay pourcentage each ETCS cause over the pourcentage each ETCS represent of the database
rate_delay_per_ETCS_Cat.index = rate_niveau_ETCS.index
Ratio_delay_per_over_ETCS_per = pd.Series(rate_niveau_ETCS/rate_delay_per_ETCS_Cat)
Ratio_delay_per_over_ETCS_per.index = rate_niveau_ETCS.index
Ratio_delay_per_over_ETCS_per
```
